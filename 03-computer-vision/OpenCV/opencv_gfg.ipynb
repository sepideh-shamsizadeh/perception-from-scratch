 This notebook is for image and video processing from the basics to advanced topics using Python and OpenCV. It will includes handle image transformations, feature extraction, object detection and more.

```python
import cv2

image = cv2.imread('road.jpg')
cv2.imshow("Image", image)
cv2.waitKey(0)
h,v = image.shape[:2]

print(f"height:{h}, weight:{v}")
```

```python
(B, G, R) = image[9,9]

print(f"RGB value at (9,9): R= {R}, G= {G}, B= {B}")


R = image[9,9,2]
print(f"R={R}")
```

```python
roi = image[100:300, 200:700]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
```

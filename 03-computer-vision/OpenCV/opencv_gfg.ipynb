```python
import cv2

image = cv2.imread('road.jpg')
cv2.imshow("Image", image)
cv2.waitKey(0)
h,w = image.shape[:2]

print(f"height:{h}, weight:{w}")
```

```python
(B, G, R) = image[9,9]

print(f"RGB value at (9,9): R= {R}, G= {G}, B= {B}")


R = image[9,9,2]
print(f"R={R}")
```

```python
roi = image[100:300, 200:700]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
```

```python
resize = cv2.resize(image, (700,700))
cv2.imshow("Resize", resize)
cv2.waitKey(0)
```

The aspect ratio of the image is not maintained. So we need to do some extra work in order to maintain a proper aspect ratio.

```python
import cv2


ratio = 1200 / w
dim = (1200, int(h * ratio))

resize_aspect = cv2.resize(image, dim)
cv2.imshow("Resized Image", resize_aspect)
cv2.waitKey(0)


# rotate resized image
h2, w2 = resize_aspect.shape[:2]
center = (w2 // 2, h2 // 2)

rot_mat = cv2.getRotationMatrix2D(center, 15, 1.0)

rotated = cv2.warpAffine(resize_aspect, rot_mat, (w2, h2))

cv2.imshow("Rotated Image", rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# cv2.rectangle(image, start_point:top left, end_point:bottom right, color, thickness)
output = image.copy()
rectangle = cv2.rectangle(output, (100,200), (400, 700), (255,0,0), 2)
cv2.imshow('Image', rectangle)
cv2.waitKey(0) # Waits indefinitely until a key is pressed.
cv2.destroyAllWindows() # Closes all windows opened by program.
```

```python
# cv2.putText(image, text, org, font, fontScale, color, thickness=1, lineType=cv2.LINE_8, bottomLeftOrigin=False)

text = "Hello, OpenCV!"
img = cv2.putText(rectangle, text, (100, 200), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 1)
cv2.imshow("Output", img)
cv2.waitKey(0)
```

```python
RGB_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
HSV_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
GRAY_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

cv2.imshow("image",img)
cv2.waitKey(0)

cv2.imshow("RGB",RGB_image)
cv2.waitKey(0)

cv2.imshow("HSV",HSV_image)
cv2.waitKey(0)

cv2.imshow("GRAY",GRAY_image)
cv2.waitKey(0)

cv2.imshow("LAB",image_lab)
cv2.waitKey(0)

```

mask = cv2.inRange(hsv_image, lower_color, upper_color)

If the pixel is inside the range → mask value becomes 255 white.
If the pixel is outside the range → mask value becomes 0 black.

So the result is a black-and-white image.

cv2.bitwise_and()

cv2.bitwise_and() applies the mask to the original image.

Example:

result = cv2.bitwise_and(image, image, mask=mask)

It means:

“Keep the pixels where the mask is white, and make the rest black.”

So if the mask found blue areas, the result will show only the blue parts of the original image.

```python
import numpy as np
image_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 100])
upper_red = np.array([10, 255, 255])

mask = cv2.inRange(image_hsv, lower_red, upper_red)

result = cv2.bitwise_and(image, image, mask=mask)

cv2.imshow("result", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
